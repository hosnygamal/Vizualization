```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('Education.csv')
```

```python
#Data Cleaning
df.College.unique()
df.dropna(inplace=True) # Drop null values
df.isnull().sum()
df.rename({'Academic Year':'Academic_Year'},axis=1 , inplace=True) # Rename column 
df.info()
```

## (Distribution of Math Test)

> Math degree distribution have taild-distribution roughly in 600 degree and a few between 300 and 400 degrees

```python
bins = np.arange(292 , df['Math'].max()+4,4) # get bins to plot
plt.hist(df['Math'],bins=bins); #hist for distrubtion one numeric variable

#Set axis X Values
plt.xticks([100,200,300,400,500,600,700,800,900],[100,200,300,400,500,600,700,800,900]);

plt.xlabel('Math Degree'); #Set x label
plt.ylabel('Count'); #set Y label
plt.title('Math Test'); # set title
```

## (Verbal Test)

> Verbal degree distribution have taild-distribution roughly in 530 degree and a few between 300 and 400 degrees

```python
bins = np.arange(305 , df['Verbal'].max()+4,4) #set bins

plt.hist(df['Verbal'],bins=bins); #hist for distrubtion one numeric variable

#set x values
plt.xticks([100,200,300,400,500,600,700,800,900],[100,200,300,400,500,600,700,800,900]);

plt.xlabel('Verbal Degree'); #set x label
plt.ylabel('Count'); #set y label
plt.title('Verbal Test'); #set title
```

## (Writing Test)
> Writing degree distribution have taild-distribution roughly in 550 degree and a few fill in between 300 and 400 degrees


```python
bins = np.arange(312 , df['Writing'].max()+4,4) #set bins

plt.hist(df['Writing'],bins=bins); #hist for distrubtion one numeric variable

plt.xticks([100,200,300,400,500,600,700,800,900],[100,200,300,400,500,600,700,800,900]);
plt.xlabel('Writing Degree');
plt.ylabel('Count');
plt.title('Writing Test');
```

## (Enrollers Genders)
> Enrolles women are higher than men

```python
plt.figure(figsize=(8,5))

ax = sb.countplot(data = df, x = 'College', hue = 'Gender')
ax.legend(loc = 0, framealpha = 1); #loc legend in right top corner with no alpha 
plt.xticks(rotation = 20); #Rotation axis X
plt.legend(title='Gender');
plt.title('Enrollers genders at colleges');
```

## (College & Tests)
> College against tests, Music enrollers do well at verbal test while engineers good at math and all colleges above 500 in writing test

```python
plt.figure(figsize=(17.7,10.27)) #set size Vizualization
c1 , c2 , c3 ,c4= sb.color_palette()[3] , sb.color_palette()[2] ,sb.color_palette()[1],sb.color_palette()[3] #Some Colors to used in plots

#First Plot
plt.subplot(131 ,frameon=False) # 
sb.boxplot(data=df , y='College' ,orient='h' ,  x='Math' , color=c1);


plt.suptitle('Colleges and tests', fontsize=14 , y=1.0, weight='bold') #Main Title

#Second Plot
plt.subplot(132 ,frameon=False)
sb.boxplot(data=df , y='College' , orient='h',x='Verbal' , color=c2);
plt.yticks([],rotation=0) #Remove axis Y and zero Rotation
plt.ylabel('') #No label Name in axis Y.

#third Plot
plt.subplot(133 ,frameon=False)
sb.boxplot(data=df , y='College' ,orient='h', x='Writing' , color=c3);
plt.yticks([],rotation=0)
plt.ylabel('')

plt.tight_layout()
```

## (Distribution Ethnicity)
> Most enrolles from White ethnicity 

```python
counts = df.groupby(['College' , 'Ethnicity']).size() #Group unique college and ethnicity
counts = counts.reset_index(name='count') #Rename Size with Column called Count
counts = counts.pivot(index='College' , columns='Ethnicity'  , values='count') #set Pivot table
#annot to show values I used fmt='d' to get int number
sb.heatmap(counts , annot=True ,cmap='rocket_r', fmt='d'); 

plt.xticks(rotation=30); #Rotation axis X
plt.title('Distribution Ethnicity'); #set title
```
